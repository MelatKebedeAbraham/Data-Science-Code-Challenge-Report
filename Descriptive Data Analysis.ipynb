```python
#Read the Dataset

import pandas as pd

file = '/content/drive/MyDrive/Data Science Data Challenge/CHAMPS(Processed).csv'
data = pd.read_csv(file)

data.head()
```

What are magnitued and proportion of each of the infant underlying casues for child death?

```python
#Step 1: Filter infant cases
infant_data = data[data['Case Type'] == 'Infant']
```

```python
#Step 2: Calculate the count (magnitude) of each unique underlying cause
condition_counts = infant_data['Condition'].value_counts()
```

```python
#Step 3: Calculate the proportion of each underlying cause
condition_proportions = condition_counts / condition_counts.sum()
```

```python
#Combine the counts and proportions into a single DataFrame
result = pd.DataFrame({
    'Magnitude': condition_counts,
    'Proportion': condition_proportions
})

# Print the result
print(result)
```

What are the proportion and magnitude of the maternal factors contributing for child death?

```python
#Step 1: Filter the DataFrame
maternal_disease_data = data.dropna(subset=['Matrtnal Disease'])
```

```python
#Step 2: Calculate the count (magnitude) of each unique maternal disease condition
maternal_disease_counts = maternal_disease_data['Matrtnal Disease'].value_counts()
```

```python
#Step 3: Calculate the proportion of each maternal disease condition
maternal_disease_proportions = maternal_disease_counts / maternal_disease_counts.sum()
```

```python
#Combine the counts and proportions into a single DataFrame
result = pd.DataFrame({
    'Magnitude': maternal_disease_counts,
    'Proportion': maternal_disease_proportions
})

# Print the result
print(result)
```
