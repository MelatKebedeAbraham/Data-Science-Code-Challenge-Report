```python
#Read the Dataset

import pandas as pd

file = '/content/drive/MyDrive/Data Science Data Challenge/CHAMPS(Processed).csv'
data = pd.read_csv(file)

data.head()
```

```python
import matplotlib.pyplot as plt

# Get the top five causes
top_five_causes = data['Condition'].value_counts().nlargest(5)

# Plot
plt.figure(figsize=(10, 6))
top_five_causes.plot(kind='bar')
plt.title('Top Five Infant Underlying Causes of Child Death')
plt.xlabel('Underlying Cause')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the top five maternal factors
top_five_factors = data['Matrtnal Disease'].value_counts().nlargest(5)

# Plot
plt.figure(figsize=(10, 6))
top_five_factors.plot(kind='bar')
plt.title('Top Five Maternal Factors Contributing to Child Death')
plt.xlabel('Maternal Factor')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the frequency of each case type
case_type_counts = data['Case Type'].value_counts()

# Plot
plt.figure(figsize=(10, 6))
case_type_counts.plot(kind='bar')
plt.title('Child Deaths Based on Case Types')
plt.xlabel('Case Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
